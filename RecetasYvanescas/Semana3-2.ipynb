```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def dens(x, y):
    return np.exp(-0.5*((x**2) + (y**2)/9 -x*y/12))


N=100000
# Genera puntos sobre una grid
x_line = np.linspace(-5,5,100)
y_line = np.linspace(-5,5,100)
x_grid, y_grid = np.meshgrid(x_line, y_line)
z_grid = dens(x_grid, y_grid)

x_lista = [np.random.random()*np.pi]
y_lista = [np.random.random()*np.pi]
sigma_delta = 1.0

# Genera listas de puntos
for i in range(1,N):
    propuestax  = x_lista[i-1] + np.random.normal(loc=0.0, scale=sigma_delta)
    propuestay  = y_lista[i-1] + np.random.normal(loc=0.0, scale=sigma_delta)
    r = min(1,dens(propuestax,propuestay)/dens(x_lista[i-1],y_lista[i-1]))
    alpha = np.random.random()
    if(alpha<r):
        x_lista.append(propuestax)
        y_lista.append(propuestay)
    else:
         x_lista.append(propuestax)
        y_lista.append(propuestay)



fig, (ax0, ax1) = plt.subplots(1,2)

# grafica los puntos de la grid
im = ax0.pcolormesh(x_grid, y_grid, z_grid)

# grafica el histograma bidimensional a partir de la lista de puntos
_ = plt.hist2d(x_lista, y_lista, bins=50)
```
